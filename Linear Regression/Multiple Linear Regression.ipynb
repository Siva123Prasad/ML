# Multiple Linear Regression: House Price Prediction

This notebook demonstrates how to implement **multiple linear regression** on a house dataset. We'll handle categorical features using **dummy variables**, fit the model, and evaluate it using RÂ² and RMSE on both train and test data.

---

## Step 1: Import Libraries


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

## Step 2: Load Dataset


```python
df = pd.read_csv("house.csv")
df.head()
```

## Step 3: Handle Categorical Variables
We'll use `pd.get_dummies` to convert categorical columns to numeric using one-hot encoding.


```python
df_encoded = pd.get_dummies(df, columns=['Material', 'Locality'], drop_first=True,dtype=int)
df_encoded.head()
```

## Step 4: Define Features and Target
We'll use all columns except `Price` as predictors.


```python
X = df_encoded.drop('Price', axis=1)
y = df_encoded['Price']
```

## Step 5: Split Data into Train and Test


```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

## Step 6: Train the Linear Regression Model


```python
model = LinearRegression()
model.fit(X_train, y_train)
```

## Step 7: Evaluate Model on Train and Test Data


```python
# Train Performance
y_train_pred = model.predict(X_train)
print("Train RÂ²:", r2_score(y_train, y_train_pred))
print("Train RMSE:", np.sqrt(mean_squared_error(y_train, y_train_pred)))

# Test Performance
y_test_pred = model.predict(X_test)
print("Test RÂ²:", r2_score(y_test, y_test_pred))
print("Test RMSE:", np.sqrt(mean_squared_error(y_test, y_test_pred)))
```

### Model Performance Evaluation

Interpretation
RÂ² Score
Train RÂ² (0.81): The model explains about 81% of the variance in the target variable on the training dataset.

Test RÂ² (0.78): On unseen data (test set), the model still captures approximately 78% of the variance.

What this Means:

Both RÂ² values are high, with only a small drop from train to test. This indicates the model is fitting the data well and generalizing effectively to new, unseen data.

There is no sign of significant overfitting; the model maintains similar predictive strength on both train and test sets.

RMSE (Root Mean Squared Error)
Train RMSE (11,963.53) / Test RMSE (11,691.08):

The RMSE is a measure of prediction errorâhow much (on average) predictions deviate from the actual target values.

The training and test errors are similar and relatively low, suggesting consistent model performance.

Lower RMSE means the model's predictions are close to actual values, in the same units as the target variable.

### Summary
The model performs strongly both on training and test data, with good predictive accuracy and no major generalization issues.

High RÂ² and similar RMSE values across train and test splits are desirable and show the model's robustness.

This level of performance indicates that the features and model selection are suitable for the prediction task at hand.

## Step 8: Visualize Actual vs Predicted on Test Data


```python
plt.scatter(y_test, y_test_pred, color='green')
plt.plot([y.min(), y.max()], [y.min(), y.max()], color='red', linestyle='--')
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Actual vs Predicted Prices (Test Set)')
plt.grid(True)
plt.show()
```

## Step 9: Interpret Model Coefficients


```python
coefficients = pd.DataFrame({
    'Feature': X.columns,
    'Coefficient': model.coef_
})
print(coefficients)
```

### The provided coefficients represent how much each feature contributes to the predicted value of the target (e.g., house price) in a linear regression model, holding all other features constant. 

Feature Effects on house price is as follows:

Each additional unit of area increases the prediction by 40.40.

Each extra bedroom adds 2,197.71 to the result.

Each additional bathroom adds 5,460.77.

Houses built with masonry add 17,739.53 compared to the baseline material.

Being in 'Riverside' locality adds 6,073.72 vs. baseline locality.

Being in 'Summit View' locality adds 35,022.21 vs. baseline locality.


### Key Takeaways
Area: A strong, positive linear effect; the price or target increases steadily with area.

Bedrooms & Bathrooms: More bedrooms or bathrooms markedly increase the target value, with bathrooms contributing more per unit than bedrooms.

Material_Masonry: Indicates a significant price (or target) premium for masonry properties compared to the reference material.

Locality Effects: Location is extremely influential. 'Summit View' stands out as the location with the highest premium, far more than 'Riverside'.
