## Twitter Sentiment Analysis

````python
import pandas as pd
import numpy as np
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import pickle

# --- Part 1: Data Processing ---

# 1. Load the Dataset
print("--- 1. Loading Dataset ---")
df_raw = pd.read_csv('twitter_training.csv', encoding='latin-1')

# 2. Print the first few rows to "visualize" the data
print(df_raw.head())
````

````python
# 3. Print the column names
print("\n--- Column Names ---")
print(df_raw.columns.tolist())
````

````python
# Adding column names for easier processing and understanding 
COLUMNS = ['tweet_id', 'subject', 'sentiment', 'text']
df = pd.read_csv('twitter_training.csv', header=None, names=COLUMNS, encoding='latin-1')
print(f"Initial DataFrame shape: {df.shape}")
print(df.head())
````

````python
print("\n--- 2. Data Cleaning ---")

# Check for and handle missing values
print("\nMissing values before cleaning:")
print(df.isnull().sum())
````

````python
df.dropna(subset=['text'], inplace=True)
print(f"DataFrame shape after dropping missing text: {df.shape}")
````

````python
# Remove duplicates
print(f"Number of duplicates before removal: {df.duplicated().sum()}")
df.drop_duplicates(inplace=True)
print(f"DataFrame shape after dropping duplicates: {df.shape}")
````

````python
def clean_text(text):
    # Remove URLs
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    # Remove mentions (@...)
    text = re.sub(r'@\w+', '', text)
    # Remove hashtags (#...) - keeping the text, removing the '#'
    text = re.sub(r'#', '', text)
    # Remove special characters and punctuation (keep only letters and spaces)
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    # Remove extra whitespace
    text = re.sub(r'\s+', ' ', text).strip()
    return text

df['cleaned_text'] = df['text'].apply(clean_text)
````

````python
import nltk

# Download the 'stopwords' resource. 
# “Stopwords” are very common words like “the”, “is”, “and” that usually don’t carry useful meaning for many NLP tasks.
# They can be often removed during preprocessing.
nltk.download('stopwords')

# Download the 'wordnet' resource, which is needed for lemmatization later
# WordNet contains information about word lemmas, synonyms, parts of speech, etc.
# This is done to standardize various similar words like 'clicked', 'clicks' to 'click'
nltk.download('wordnet')
# Tokenization, Lowercasing, Stop Word Removal, and word standardization
lemma = WordNetLemmatizer()
sentiment_stopwords = set(stopwords.words('english')) - {'not', 'no', 'nor', 'never', 'neither', 'nobody', 'nothing', 'nowhere', 'hardly', 'barely'}

def preprocess_text(text):
    # Tokenize and convert to lowercase
    tokens = text.lower().split()
    # Remove stop words and lemmatize
    tokens = [lemma.lemmatize(word) for word in tokens if word not in stop_words]
    return ' '.join(tokens)

df['processed_text'] = df['cleaned_text'].apply(preprocess_text)
df['processed_text'] = df['processed_text'].astype(str).fillna('') # Handle potential NaNs

# Filter out 'Irrelevant' sentiment as per the model's 3-class requirement
df = df[df['sentiment'].isin(['Positive', 'Negative', 'Neutral'])].copy()


# Remove empty texts
df = df[df['processed_text'].str.strip() != ''].copy()
print(f"Final dataset size: {df.shape}")

# Sample check
print("\nSample processed texts:")
for sent in ['Negative', 'Neutral', 'Positive']:
    sample = df[df['sentiment'] == sent]['processed_text'].iloc[0]
    print(f"  {sent}: '{sample}'")
````

````python
print("\n--- Now let's make model understand what the input is. Let's bucket each sentiment together and assign values ---")

# Convert sentiment labels to numerical format
sentiment_map = {'Negative': 0, 'Neutral': 1, 'Positive': 2}
df['sentiment_label'] = df['sentiment'].map(sentiment_map).astype(int)

# Create a sequence of tokenized words for each tweet
MAX_WORDS = 10000
MAX_LEN = 50

tokenizer = Tokenizer(num_words=MAX_WORDS, oov_token="<oov>")
tokenizer.fit_on_texts(df['processed_text'])

sequences = tokenizer.texts_to_sequences(df['processed_text'])
X = pad_sequences(sequences, maxlen=MAX_LEN, padding='post', truncating='post')
y_raw = df['sentiment_label'].values

print(f"Vocabulary size: {len(tokenizer.word_index)}")
print(f"Shape of padded sequences (features): {X.shape}")
print(f"Shape of labels: {y_raw.shape}")

print(f"\nClass distribution:")
print(pd.Series(y_raw).value_counts().sort_index())
````

````python
print("\n\n--- Let's understand more about the actual tweets and find the sentiment count ---")

# 1. Basic Statistics
sentiment_counts = df['sentiment'].value_counts()
print("\nDistribution of Tweet Sentiments:")
print(sentiment_counts)

df['tweet_length'] = df['processed_text'].apply(lambda x: len(str(x).split()))
print("\nSummary Statistics for Tweet Length:")
print(df['tweet_length'].describe())
````

````python
from collections import Counter
# Frequency of top words in each sentiment
def get_top_n_words(corpus, n=20):
    words = ' '.join(corpus.astype(str)).split()
    counter = Counter(words)
    most_common = counter.most_common(n)
    return dict(most_common)

top_words = {}
for sentiment in df['sentiment'].unique():
    corpus = df[df['sentiment'] == sentiment]['processed_text']
    top_words[sentiment] = get_top_n_words(corpus)

fig, axes = plt.subplots(1, 3, figsize=(18, 6))
fig.suptitle('Top 20 Most Frequent Words by Sentiment (After Preprocessing)', fontsize=16)

sentiment_order = ['Negative', 'Neutral', 'Positive']
palette_map = {'Negative': "Reds", 'Neutral': "Blues", 'Positive': "Greens"}

for i, sentiment in enumerate(sentiment_order):
    words = top_words.get(sentiment, {})
    ax = axes[i]
    if words:
        sns.barplot(x=list(words.values()), y=list(words.keys()), ax=ax, palette=palette_map[sentiment])
        ax.set_title(f'{sentiment} Sentiment')
        ax.set_xlabel('Frequency')
        ax.set_ylabel('Word')
    else:
        ax.set_title(f'{sentiment} Sentiment (No data)')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
````

# Model training explanation

### Embedding layer: Map each integer token ID to a 100-dimensional dense vector. This learns semantic relationships—words like "great" and "awesome" end up close together in vector space; "terrible" and "awful" cluster separately.

#### Layer 1: Embedding

Input: Sequence of word IDs like [42, 17, 893, 0, 0, ...] (length 50)

Output: Sequence of 100-dimensional vectors

What it does: Looks up each word ID in a learned "dictionary" that maps words to meaningful vectors. Words like "great" and "awesome" end up with similar vectors.

### LSTM Layers: Process the sequence of word vectors in order, capturing contextual patterns. Example: "This product is not good" (negative context) vs. "Not bad at all" (positive despite "not"). Two stacked LSTM layers learn increasingly abstract patterns.

#### Layer 2: Bidirectional LSTM

Input: Sequence of 50 word vectors

Output: Single 128-dimensional vector (64×2 for both directions)

What it does: Reads the sequence forwards AND backwards, capturing context from both sides. Understands that "not good" is different from "good".

### Classification layers: Dense layers + softmax output produce probabilities for the three sentiment classes.

#### Layer 3: Dropout (0.5)

What it does: During training, randomly "turns off" 50% of neurons each batch

Why: Prevents overfitting by forcing the network to not rely on any single neuron too much. Makes the model more robust.


#### Layer 4: Dense (32 units, ReLU)

Input: 128-dimensional vector

Output: 32-dimensional vector

What it does: A fully connected layer that combines features into higher-level patterns. ReLU activation keeps only positive values.

Why: Adds capacity to learn complex decision boundaries between sentiment classes.

#### Layer 5: Dense (3 units, Softmax)

Input: 32-dimensional vector

Output: 3 probabilities that sum to 1.0

What it does: Final classification. Outputs something like [0.1, 0.2, 0.7] meaning 10% Negative, 20% Neutral, 70% Positive.

Why: Softmax ensures we get valid probabilities for all three classes.

### Now that we have built the model, let's see how it performs in the real world. 

````python

print("\n\n--- Sample Predictions ---")

def predict_sentiment(text):
    """
    Predict sentiment of a given tweet text.
    
    Args:
        text (str): Raw tweet text
    
    Returns:
        dict: Contains prediction, probability, and sentiment label
    """
    # 1. Clean the text using same preprocessing as training
    cleaned = clean_text(text)
    
    # 2. Preprocess (remove stopwords, lemmatize)
    tokens = cleaned.lower().split()
    processed = ' '.join([lemma.lemmatize(word) for word in tokens 
                         if word not in sentiment_stopwords and len(word) > 1])
    
    # 3. Tokenize using the same tokenizer from training
    sequence = tokenizer.texts_to_sequences([processed])
    
    # 4. Pad to same length as training
    padded = pad_sequences(sequence, maxlen=MAX_LEN, padding='post', truncating='post')
    
    # 5. Get prediction from model
    prediction = model.predict(padded, verbose=0)
    predicted_class = np.argmax(prediction)
    confidence = prediction[0][predicted_class]
    
    # 6. Map back to sentiment label
    sentiment_map_reverse = {0: 'Negative', 1: 'Neutral', 2: 'Positive'}
    predicted_sentiment = sentiment_map_reverse[predicted_class]
    
    return {
        'text': text,
        'cleaned_text': processed,
        'predicted_sentiment': predicted_sentiment,
        'confidence': float(confidence),
        'probabilities': {
            'Negative': float(prediction[0][0]),
            'Neutral': float(prediction[0][1]),
            'Positive': float(prediction[0][2])
        }
    }

# Test on sample tweets
print("\nSample Predictions on Different Tweets:\n")

sample_tweets = [
    "I love this product! It's amazing and works perfectly!",
    "This is terrible. Complete waste of money. Very disappointed.",
    "The product is okay. Nothing special but not bad either.",
    "not good at all, worst purchase ever",
    "absolutely fantastic experience highly recommend"
]

for tweet in sample_tweets:
    result = predict_sentiment(tweet)
    print(f"Tweet: {result['text']}")
    print(f"Cleaned: {result['cleaned_text']}")
    print(f"Prediction: {result['predicted_sentiment']} (Confidence: {result['confidence']:.2%})")
    print(f"Probabilities: {result['probabilities']}")
    print("-" * 80)
````

## Hyperparameter Tuning 
Hyperparameters are settings you choose before training begins—they control how the model learns, not what it learns. The model cannot learn these values from data; you must set them manually or search for optimal values.


### Hyperparameters Being Tuned in this Grid Search
We're tuning 3 hyperparameters:

#### Batch Size (batch_size: [32, 64])
What it is: Number of training samples the model sees before updating weights.

| Value | Effect |
|--------|--------|
**32 (smaller)**	| More frequent weight updates → Noisier but can escape local minima → Slower training
**64 (larger)**	| Fewer updates → Smoother gradients → Faster training but may miss fine patterns


#### Number of Epochs (epochs: [10, 15])
What it is: How many times the model sees the entire training dataset.
| Value | Effect |
|--------|--------|
**10 (fewer)**	| Less training time → May underfit (not learn enough)
**15 (more)**	| More training time → May overfit (memorize training data)


#### Learning Rate (learning_rate: [0.001, 0.0005])
What it is: How big each weight adjustment step is during training.

| Value | Effect |
|--------|--------|
**0.001 (higher)**	| Larger steps → Learns faster but may overshoot optimal values
**0.0005 (lower)**	| Smaller steps → Learns slower but more precise convergence



````python
# ---  Hyperparameter Tuning with Grid Search ---
print("\n\n--- Part 4B: Hyperparameter Tuning ---")

from sklearn.model_selection import ParameterGrid

# Define hyperparameter grid
param_grid = {
    'batch_size': [32, 64],
    'epochs': [10, 15],
    'learning_rate': [0.001, 0.0005]
}

# Note: Full grid search on Keras models is complex with GridSearchCV
# We'll implement a simpler manual grid search

print(f"Testing {len(list(ParameterGrid(param_grid)))} parameter combinations...\n")

results = []

for params in ParameterGrid(param_grid):
    print(f"Testing: batch_size={params['batch_size']}, "
          f"epochs={params['epochs']}, lr={params['learning_rate']}")
    
    # Build model with current parameters
    model_test = Sequential([
        Embedding(input_dim=MAX_WORDS, output_dim=EMBEDDING_DIM),
        Bidirectional(LSTM(64)),
        Dropout(0.5),
        Dense(32, activation='relu'),
        Dense(NUM_CLASSES, activation='softmax')
    ])
    
    # Compile with current learning rate
    from tensorflow.keras.optimizers import Adam
    model_test.compile(
        optimizer=Adam(learning_rate=params['learning_rate']),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    
    # Train with current batch size and epochs
    history_test = model_test.fit(
        X_train, y_train,
        epochs=params['epochs'],
        batch_size=params['batch_size'],
        validation_split=0.1,
        verbose=0
    )
    
    # Evaluate on test set
    loss, accuracy = model_test.evaluate(X_test, y_test, verbose=0)
    
    # Store results
    results.append({
        'batch_size': params['batch_size'],
        'epochs': params['epochs'],
        'learning_rate': params['learning_rate'],
        'test_accuracy': accuracy,
        'test_loss': loss,
        'final_train_acc': history_test.history['accuracy'][-1],
        'final_val_acc': history_test.history['val_accuracy'][-1]
    })
    
    print(f"  → Test Accuracy: {accuracy:.4f}, Test Loss: {loss:.4f}\n")

# Convert results to DataFrame for easy analysis
results_df = pd.DataFrame(results)
print("\n--- Grid Search Results Summary ---")
print(results_df.to_string(index=False))

# Find best parameters
best_idx = results_df['test_accuracy'].idxmax()
best_params = results_df.iloc[best_idx]

print(f"\n--- Best Parameters ---")
print(f"Batch Size: {int(best_params['batch_size'])}")
print(f"Epochs: {int(best_params['epochs'])}")
print(f"Learning Rate: {best_params['learning_rate']}")
print(f"Test Accuracy: {best_params['test_accuracy']:.4f}")
print(f"Test Loss: {best_params['test_loss']:.4f}")

# Visualize grid search results
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
for lr in results_df['learning_rate'].unique():
    subset = results_df[results_df['learning_rate'] == lr]
    plt.plot(subset['batch_size'], subset['test_accuracy'], 
             marker='o', label=f'LR={lr}')
plt.xlabel('Batch Size')
plt.ylabel('Test Accuracy')
plt.title('Grid Search: Batch Size vs Accuracy')
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
for bs in results_df['batch_size'].unique():
    subset = results_df[results_df['batch_size'] == bs]
    plt.plot(subset['learning_rate'], subset['test_accuracy'], 
             marker='s', label=f'BS={int(bs)}')
plt.xlabel('Learning Rate')
plt.ylabel('Test Accuracy')
plt.title('Grid Search: Learning Rate vs Accuracy')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
````
